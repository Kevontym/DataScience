```python
base_path = "/content/drive/MyDrive/pubmd/pqaa"
all_train_data = []

# Loop through all 12 folds
for fold in range(12):
    folder_name = f"pqal_fold{fold}"  # pqal_fold0, pqal_fold1, ..., pqal_fold11
    train_file_path = os.path.join(base_path, folder_name, "train_set.json")

    if os.path.exists(train_file_path):
        with open(train_file_path, "r") as f:
            train_data = json.load(f)  # Load JSON as dictionary
            all_train_data.extend(train_data.values())  # Extract values and append to list

# Convert to Pandas DataFrame
df_train = pd.DataFrame(all_train_data)
df_train.head()
```

```python
print(df_train.shape)  # Should show (large_number, 3) if successful
print(df_train.columns)  # Should include ['question', 'context', 'final_decision']
```

```python
!pip install transformers datasets
from transformers import BertTokenizer
import torch
```

```python
tokenizer = BertTokenizer.from_pretrained("dmis-lab/biobert-v1.1")
```

```python
label_map = {"yes": 1, "no": 0, "maybe": 2}
df_train["LABELS"] = df_train["final_decision"].map(label_map)
```

```python
def tokenize_function(examples):
    return tokenizer(
        examples["QUESTION"], examples["CONTEXTS"],
        max_length=512, truncation=True, padding="max_length"
    )

# Apply tokenization
tokenized_data = df_train.apply(tokenize_function, axis=1)

# Convert to Tensor format
input_ids = torch.tensor([x["input_ids"] for x in tokenized_data])
attention_mask = torch.tensor([x["attention_mask"] for x in tokenized_data])
labels = torch.tensor(df_train["LABELS"].tolist())
```

```python
from torch.utils.data import TensorDataset

train_dataset = TensorDataset(input_ids, attention_mask, labels)
print(f"Dataset Ready: {len(train_dataset)} samples")
```

```python
import torch
from torch.utils.data import DataLoader, RandomSampler
from transformers import BertForSequenceClassification, AdamW, get_scheduler
```

```python
batch_size = 8  # Adjust based on your GPU memory

train_dataloader = DataLoader(
    train_dataset,
    sampler=RandomSampler(train_dataset),
    batch_size=batch_size
)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = BertForSequenceClassification.from_pretrained(
    "dmis-lab/biobert-v1.1", num_labels=3
)
model.to(device)
```

```python
optimizer = AdamW(model.parameters(), lr=2e-5, eps=1e-8)

# Define learning rate scheduler
epochs = 3  # Fine-tuning typically requires 3-5 epochs
num_training_steps = len(train_dataloader) * epochs
lr_scheduler = get_scheduler(
    name="linear",
    optimizer=optimizer,
    num_warmup_steps=0,
    num_training_steps=num_training_steps
)
```

```python
import torch.nn as nn
from tqdm import tqdm

loss_fn = nn.CrossEntropyLoss()

for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")

    model.train()
    total_loss = 0

    loop = tqdm(train_dataloader, leave=True)
    for batch in loop:
        batch = tuple(t.to(device) for t in batch)  # Move batch to GPU

        input_ids, attention_mask, labels = batch

        optimizer.zero_grad()

        outputs = model(input_ids, attention_mask=attention_mask)
        loss = loss_fn(outputs.logits, labels)

        loss.backward()
        optimizer.step()
        lr_scheduler.step()

        total_loss += loss.item()
        loop.set_description(f"Loss: {total_loss / len(train_dataloader):.4f}")

    print(f"Epoch {epoch + 1} Completed ✅ | Loss: {total_loss:.4f}")
```

```python
# Load Validation Data
val_data = []
val_base_path = "/content/drive/MyDrive/pubmd/pqaa"

for fold in range(12):
    folder_name = f"pqal_fold{fold}"
    val_file_path = os.path.join(val_base_path, folder_name, "dev_set.json")

    if os.path.exists(val_file_path):
        with open(val_file_path, "r") as f:
            val_json = json.load(f)
            val_data.extend(val_json.values())

# Convert to DataFrame
df_val = pd.DataFrame(val_data)

# Tokenize validation data
val_tokenized = df_val.apply(tokenize_function, axis=1)

# Convert to tensors
val_input_ids = torch.tensor([x["input_ids"] for x in val_tokenized])
val_attention_mask = torch.tensor([x["attention_mask"] for x in val_tokenized])
val_labels = torch.tensor(df_val["final_decision"].map(label_map).tolist())

# Create PyTorch dataset
val_dataset = TensorDataset(val_input_ids, val_attention_mask, val_labels)
val_dataloader = DataLoader(val_dataset, batch_size=8)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Set model to evaluation mode
model.eval()

all_preds = []
all_labels = []

with torch.no_grad():
    for batch in val_dataloader:
        batch = tuple(t.to(device) for t in batch)
        input_ids, attention_mask, labels = batch

        outputs = model(input_ids, attention_mask=attention_mask)
        preds = torch.argmax(outputs.logits, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Compute accuracy
accuracy = accuracy_score(all_labels, all_preds)
print(f"Validation Accuracy: {accuracy:.4f}")

# Print classification report
print(classification_report(all_labels, all_preds, target_names=["No", "Yes", "Maybe"]))
```

```python
model.save_pretrained("biobert_pubmedqa")
tokenizer.save_pretrained("biobert_pubmedqa")
```

```python
def chatbot_response(question, context):
    model.eval()

    # Tokenize input
    inputs = tokenizer(question, context, return_tensors="pt", truncation=True, max_length=512, padding="max_length")
    inputs = {key: val.to(device) for key, val in inputs.items()}

    # Get model prediction
    with torch.no_grad():
        output = model(**inputs)
        pred_label = torch.argmax(output.logits, dim=1).item()

    # Convert back to human-readable format
    label_map_reverse = {0: "No", 1: "Yes", 2: "Maybe"}
    return label_map_reverse[pred_label]
```

```python
new_question = "Can vitamin C boost immune system function?"
new_context = "Studies show vitamin C can reduce symptoms of colds, but its effect on the immune system is debated."

response = chatbot_response(new_question, new_context)
print(f"Chatbot Response: {response}")
```
